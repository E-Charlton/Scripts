```python
import numpy as np
from numpy import pi
import matplotlib
from matplotlib import pyplot
import matplotlib.pyplot as plt
```

All information used is from Narayan 1992 (Narayan, Ramesh. "The Physics of Pulsar Scintillation." Philosophical Transactions: Physical Sciences and Engineering 341, no. 1660 (1992): 151-65. http://www.jstor.org/stable/53917.)

Here I have recreated Figure 1 from the text.

```python
rd=np.arange(0.01,100,0.01)
rf=np.arange(0.01,100,0.01)
rat=np.arange(0.01,100,0.01)
v=1.
rref=rat*rf
tf=rf/v
ts1=tf
ts2=(rf/rat)/v
ts3=rref/v

n1=100
n2=9999

plt.plot(np.log(rat)[:100],np.log(ts1/tf)[:100], label='Weak', color='k')
plt.plot(np.log(rat[n1:n2]),np.log((ts2/tf)[n1:n2]), label='Diffractive', color='k')
plt.plot(np.log(rat[n1:n2]),np.log((ts3/tf)[n1:n2]), label='Refractive', color='k')
plt.annotate(xy=(-0.75,0.1 ), s='Weak', fontsize=12)
plt.annotate(xy=(1.2,1.4), rotation=30, s='Refractive', fontsize=12)
plt.annotate(xy=(1.2,-1.1), rotation=335, s='Diffractive', fontsize=12)
plt.xlim(-1,2)
plt.ylim(-2,2)
plt.xlabel(r"ln ($r_{\mathrm{F}}/r_{\mathrm{diff}})$", fontsize=14)
plt.ylabel(r"ln ($t_{\mathrm{scint}}/t_{\mathrm{F}})$", fontsize=14)
plt.savefig('narayan_timescale.png')
plt.show()
```

```python
rd=np.arange(0.01,100,0.01)
rf=np.arange(0.01,100,0.01)
rat=np.arange(0.01,100,0.01)
v=1.
rref=rat*rf
tf=rf/v
ts1=(rat)**(5./6.)
ts2=np.zeros(9999)+1
ts3=(1./rat)**(1./3.)

n1=100
n2=9999

plt.plot(np.log(rat)[:100],np.log(ts1)[:100], label='Weak', color='k')
plt.plot(np.log(rat[n1:n2]),np.log(ts2)[n1:n2], label='Diffractive', color='k')
plt.plot(np.log(rat[n1:n2]),np.log(ts3)[n1:n2], label='Refractive', color='k')
plt.annotate(xy=(-0.8,-0.38 ), rotation=55, s='Weak', fontsize=12)
plt.annotate(xy=(0.8, -0.09), s='Refractive', fontsize=12)
plt.annotate(xy=(0.8,-0.38), rotation=330, s='Diffractive', fontsize=12)
plt.xlim(-1,2)
plt.ylim(-1,0.2)
plt.xlabel(r"ln ($r_{\mathrm{F}}/r_{\mathrm{diff}})$", fontsize=14)
plt.ylabel(r"ln $m_{\mathrm{p}}$", fontsize=14)
plt.savefig('narayan_modscale.png')
plt.show()
```
